<h3>To start, we import pandas and create a dataframe out of our info from our CSV file.</h3>

```python
import pandas as pd
file = "Movie Ratings.csv"
ratings = pd.read_csv(file)
```

```python
ratings
```

<h3>Now we will get the average rating for each user.</h3>
<h5>Note that we will not include Null/NaN values in the average.</h5>

```python
person_avg = ratings.set_index("Name")
person_avg['Average'] = person_avg.mean(axis=1)
person_avg['Average']
```

<h3>Now we will get the average rating for each movie.</h3>
<h5>Note that we will not include Null/NaN values in the average.</h5>

```python
movie_avg = person_avg
movie_avg.loc["Average"] = movie_avg.mean()
movie_avg.loc["Average"]
```

<h3>Here is what the whole table looks like with the averages</h3>

```python
movie_avg
```

<h3>Now we will convert the movie averages dataframe to have normalized data.</h3>

```python
normalized = (movie_avg - movie_avg.min()) / (movie_avg.max() - movie_avg.min())
normalized
```

<h3>Now we will convert the movie averages dataframe to have standardized data.</h3>

```python
standardized = (movie_avg - movie_avg.mean())/movie_avg.std()
standardized
```
